```python
import pandas as pd

ipl_data = {'Team': ['Riders', 'Riders', 'Devils', 'Devils', 'Kings',
         'kings', 'Kings', 'Kings', 'Riders', 'Royals', 'Royals', 'Riders'],
         'Rank': [1, 2, 2, 3, 3,4 ,1 ,1,2 , 4,1,2],
         'Year': [2014,2015,2014,2015,2014,2015,2016,2017,2016,2014,2015,2017],
         'Points':[876,789,863,673,741,812,756,788,694,701,804,690]}
df = pd.DataFrame(ipl_data)

df
```

```python
df.pivot_table(index="Team")  # default mean of coloumns
```

```python
#___Yeh mene Team & Points ka  mean 

df[["Team","Points"]].pivot_table(index="Team")  # default mean of coloumns
```

```python
df.pivot_table(index="Team",columns="Year")  # Group by   Mean
```

```python
df.pivot_table(index="Team",columns="Year",aggfunc='count')  
```

```python
df.pivot_table(index="Team",columns="Year",aggfunc='sum')  
```

```python
df.pivot_table(index="Team",columns="Year",aggfunc='min')  
```

```python
df[["Team","Points","Year"]].pivot_table(index="Team",columns="Year", aggfunc="sum")  # default mean of coloumns
```

# _Nan lo filling karna fillna 


```python
#_Nan lo filling karna fillna 

df.pivot_table(index="Team",columns="Year",aggfunc='min',fill_value=0)  
```

# _margins charo tarf se sum karna 

```python
df.pivot_table(index="Team",columns="Year",aggfunc='sum',margins=True)  
```

```python
df.pivot_table(index="Team",columns="Year",aggfunc='sum',fill_value=0,margins=True)  
```
