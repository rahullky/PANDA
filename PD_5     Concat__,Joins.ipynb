```python
One = pd.DataFrame(
                {'id':[1,2,3,44,5],
                 'Name':['Billy','Brain','Bran','Betty','Brazil'],
                  'Subject_id':['sub2','sub4','sub3','sub6','sub5']}
                                                        )
```

```python
Two = pd.DataFrame(
                {'id':[11,2,3,4,5],
                 'Name':['Billy','raj','Bran','Betty','Ram'],
                  'Subject_id':['sub2','sub4','sub3','sub6','sub5']}
                                                                                         )
```

```python
dd = pd.merge(One,Two,on=['id'])
dd
```

```python
ddd = pd.merge(One,Two,on=['Name'])
ddd
```

```python
dd1 = pd.merge(One,Two,on=['id','Name'])
dd1
```

```python
dd2 = pd.merge(One,Two,on='id',how='left')
dd2
```

```python
dd3 = pd.merge(One,Two,on='id',how='right')
dd3
```

```python
dd4 = pd.merge(One,Two,on='id',how='outer')
dd4
```
